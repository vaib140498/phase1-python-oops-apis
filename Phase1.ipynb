```python
from google.colab import userdata
import requests

def get_latest_news():
  topic = ['apple', 'samsung', 'tesla']
  api_key =userdata.get('apikey')
  url = "https://newsapi.org/v2/everything"
  responses = []
  for i in range(0,len(topic)):
    params = {
      "q": topic[i],
      "from":"2026-05-11",
      "sortBy":"popularity",
      "apiKey":api_key
      }
    response = requests.get(url,params=params)
    responses.append(response)

  for response in responses:
    if response.status_code == 200:
      data = response.json()
      articles = data['articles']

      print("\nTop News\n")

      for article in articles[:5]:
        print(article['title'])
        print(article['url'])
        print()
    else:
      print("Failed to retrieve data")

get_latest_news()

```

```python
from google.colab import userdata
import requests

def get_latest_TopNews(topics):

  api_key = userdata.get('apikey')
  url = "https://newsapi.org/v2/everything"

  all_articles = []

  for topic in topics:
    params = {
        "q": topic,
        "from": "2026-05-11",
        "sortBy": "popularity",
        "apiKey": api_key
    }
    response = requests.get(url, params=params)
    if response.status_code == 200:
      data = response.json()
      articles = data['articles']
      for article in articles:
        article['search_topic'] = topic
      all_articles.extend(articles[:5])
    else:
      print(f"Failed to retrieve {topic} data : {response.status_code}")
  return all_articles
```

Day-2, creating save_news(articles,filename) as a JSON file (JavaScript Object Notation).

```python
import json

topics = ['Tesla', 'Iran War']
art = get_latest_TopNews(topics)

#Saving articles as JSON file
with open('news_data.json', 'w') as file:
  json.dump(art, file)
  print("Data saved to news_data.json")

#Loading and reading saved file
with open('news_data.json', 'r') as file:
  loaded_data = json.load(file)


print(f'\nLoaded {len(loaded_data)} articles:\n')
for article in loaded_data:
    print(f"[{article['search_topic']}] {article['title']}")
    print(f"URL: {article['url']}")
    print()
```
