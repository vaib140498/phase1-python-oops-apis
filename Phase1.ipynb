```python
from google.colab import userdata
import requests

def get_latest_TopNews(topics):

  api_key = userdata.get('apikey')
  url = "https://newsapi.org/v2/everything"

  all_articles = []

  for topic in topics:
    params = {
        "q": topic,
        "from": "2026-05-11",
        "sortBy": "popularity",
        "apiKey": api_key
    }
    response = requests.get(url, params=params)
    if response.status_code == 200:
      data = response.json()
      articles = data['articles']
      for article in articles:
        article['search_topic'] = topic
      all_articles.extend(articles[:5])
    else:
      print(f"Failed to retrieve {topic} data : {response.status_code}")
  return all_articles
```

Day-2, creating save_news(articles,filename) as a JSON file (JavaScript Object Notation).

```python
import json

topics = ['Tesla', 'Iran War']
art = get_latest_TopNews(topics)

#Saving articles as JSON file
with open('news_data.json', 'w') as file:
  json.dump(art, file)
  print("Data saved to news_data.json")

#Loading and reading saved file
with open('news_data.json', 'r') as file:
  loaded_data = json.load(file)


print(f'\nLoaded {len(loaded_data)} articles:\n')
for article in loaded_data:
    print(f"[{article['search_topic']}] {article['title']}")
    print(f"URL: {article['url']}")
    print()
```

```python
from google.colab import files
from pypdf import PdfReader
from sentence_transformers import SentenceTransformer
import re
from sklearn.metrics.pairwise import cosine_similarity
import nltk

nltk.download('punkt')
nltk.download('punkt_tab')

from nltk.tokenize import sent_tokenize



def clean_text(text):

    text = text.lower()

    text = re.sub(r'\n', ' ', text)

    text = re.sub(r'\s+', ' ', text)

    text = re.sub(r'[^a-zA-Z0-9@.,:/+() -]', '', text)

    return text.strip()



uploaded = files.upload()
pdf_name = list(uploaded.keys())[0]
reader = PdfReader(pdf_name)

text = ""

for page in reader.pages:
  text += page.extract_text()
final = clean_text(text)
sentences = sent_tokenize(final)

model = SentenceTransformer('all-MiniLM-L6-v2')

embeddings = model.encode(sentences)
chunks = []
current_chunk = [sentences[0]]

threshold = 0.7

for i in range(1, len(sentences)):

    similarity = cosine_similarity(
        [embeddings[i-1]],
        [embeddings[i]]
    )[0][0]

    if similarity >= threshold:
        current_chunk.append(sentences[i])

    else:
        chunks.append(" ".join(current_chunk))
        current_chunk = [sentences[i]]

# Add last chunk
chunks.append(" ".join(current_chunk))

# Print chunks
for i, chunk in enumerate(chunks):
    print(f"\nChunk {i+1}:")
    print(chunk)
```
