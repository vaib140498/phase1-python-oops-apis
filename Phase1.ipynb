```python
#Now learning how to use API's in python
##Installing API's 'request' ---> pip install requests
##Using 'requests'
import requests

def get_latest_news():

  url = f"https://newsapi.org/v2/everything?q=apple&from=2026-05-11&to=2026-05-11&sortBy=popularity&apiKey=d772a9169da340a78aa00be7d1d52327"

  response = requests.get(url)
  if response.status_code == 200:
    data = response.json()
    articles = data['articles']
    for article in articles:
      author = article['author']
      print(author)
  else:
    print("Failed to retrieve data")

get_latest_news()
```
