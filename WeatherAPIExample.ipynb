```python
from google.colab import userdata
import requests
import json

def get_weather(pincode):
  apikey = userdata.get('weatherAPIkey')
  url = f'http://api.weatherapi.com/v1/forecast.json'

  weather_data = []

  params = {
      'key': apikey,
      "q": pincode,
      "days": 4,
  }
  response = requests.get(url,params=params)
  if response.status_code == 200:
    weather_data = response.json()
  else:
    print(f'Error: {response.status_code}')
  return weather_data



data = get_weather("Varanasi")

with open('weather.json','w') as f:
  json.dump(data,f,indent=4)

with open('weather.json','r') as f:
  data = json.load(f)

print(data['location']['name'])
print(data['current']['temp_f'])
print(data['forecast']['forecastday'])
```

```python
from google.colab import userdata
import requests

def get_weather(pincode):
    apikey = userdata.get('weatherAPIkey')
    url = 'http://api.weatherapi.com/v1/forecast.json'

    params = {
        "q": pincode,
        "days": 5,
        "key": apikey
    }

    try:
        response = requests.get(url, params=params)
        response.raise_for_status()   # ← automatically raises error if not 200
        weather_data = response.json()

        location  = weather_data['location']['name']
        temp_c    = weather_data['current']['temp_c']
        condition = weather_data['current']['condition']['text']

        print(f"Weather in {location}: {temp_c}°C, {condition}")
        return weather_data

    except requests.exceptions.ConnectionError:
        print("No internet connection")
    except requests.exceptions.HTTPError as e:
        print(f"HTTP error: {e}")
    except KeyError as e:
        print(f"Unexpected response format, missing key: {e}")
    except Exception as e:
        print(f"Something went wrong: {e}")

    return None   # always return something, even on failure

get_weather("INVALIDPINCODE999")  # triggers HTTPError
get_weather("")                    # triggers HTTPError

def save_news(articles, filename):
    try:
        with open(filename, 'w') as f:
            json.dump(articles, f, indent=2)
        print(f"Saved {len(articles)} articles to {filename}")
    except IOError as e:
        print(f"Could not save file: {e}")
```
